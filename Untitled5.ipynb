```python
def pixel_dividing(img , n  ,b_z) : 
    x = img.shape[0]
    y = img.shape[1]
    
    # getting 8*8 pixels
    blocks = np.array([img[i:i+b_z, j:j+b_z , :] for i in range(0,x,16) for j in range(0,y,16)])
    
    # getting only one color 
    BLOCKS=blocks[n]
    print(np.shape(blocks))
    return BLOCKS
```

```python
def searching_area(img , n ) : 
    x = img.shape[0]
    y = img.shape[1]
    
    # getting 8*8 pixels
    blocks = np.array([img[(i+8)-64:(i+8)+64 , (j+8)-64:(j+8)+64 , :] for i in range(0,x,16) for j in range(0,y,16)])
    
    # getting only one color 
    BLOCKS=blocks[n]
    print(np.shape(blocks))
    return BLOCKS
```
